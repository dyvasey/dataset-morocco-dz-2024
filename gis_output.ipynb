# Code to produce files for use in GIS

```python
import pandas as pd
import geopandas as gpd

from geoscripts.dz import dz

import cmcrameri.cm as cmc
from matplotlib.colors import to_hex
```

```python
# Load all dz objects
dz_path = 'dz/'
smps = dz.load_all(dz_path)
smps.sort(key=lambda x: x.name)
print([smp.name for smp in smps])
```

```python
# Output locations
df = pd.DataFrame(columns=['Latitude','Longitude','Source','Age'])
for smp in smps:
    print(smp.name)
    print(smp.latlon)

    lat,lon = smp.latlon
    df.loc[smp.name,'Latitude'] = lat
    df.loc[smp.name,'Longitude'] = lon
    df.loc[smp.name,'Source'] = smp.source
    df.loc[smp.name,'Age'] = smp.reported_age

df.to_csv('dz_locations.csv')
```
