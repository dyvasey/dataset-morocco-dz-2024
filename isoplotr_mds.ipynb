## Create MDS plots using IsoplotR

```python
%load_ext rpy2.ipython
```

```python
%%R
library('IsoplotR')
sessionInfo()
```

```python
%%R
filenames <- dir('isoplotr/input_data',pattern='*.csv')
print(filenames)
```

```python
%%R
# Get MDS file and do MDS analysis
path <- 'isoplotr/input_data/'

dz <- read.data(file.path(path,'ages.csv'),method='detritals')
plot_path <- file.path(path,'mds_plot.pdf')

pdf(plot_path)
mds(dz,nnlines=FALSE,pch=21,cex=3)
dev.off()
```
