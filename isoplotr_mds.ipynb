## Create MDS plots using IsoplotR

```python
import pandas as pd
```

```python
%load_ext rpy2.ipython
```

```python
%%R
library('IsoplotR')
sessionInfo()
```

```python
%%R
filenames <- dir('isoplotr/input_data',pattern='*.csv')
print(filenames)
```

```python
# Combine ages CSVs in Python
path = 'isoplotr/input_data/'
lc_ages = pd.read_csv(path+'laserchron_ages.csv')
lc_ages_noNE = lc_ages.iloc[:,:-3] # Exclude NE samples

dom_ages = pd.read_csv(path+'domenech_ages.csv')

per_ages = pd.read_csv(path+'perez_ages.csv')
per_ages_Tronly = per_ages[['06MCO','07MCO','08MCO','09MCO','10MCO','11MCO','13MCO','16MCO']]

all_ages = pd.concat([lc_ages_noNE,dom_ages,per_ages_Tronly],axis=1)
all_ages.to_csv(path+'all_ages.csv',index=False)
```

```python
%%R
# Get MDS file and do MDS analysis
path <- 'isoplotr/input_data/'

dz <- read.data(file.path(path,'all_ages.csv'),method='detritals')
plot_path <- file.path(path,'mds_plot.pdf')

pdf(plot_path)
mds(dz,nnlines=FALSE,pch=21,cex=3)
dev.off()
```
