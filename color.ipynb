```python
import numpy as np
import glasbey
import seaborn as sns
from cmcrameri import cm
```

```python
# DZ Spans

cmap = cm.batlowS

colors = cmap.colors[:5]

hex_samples = ['#{:02x}{:02x}{:02x}'.format(*( (c*255).astype(int) )) for c in colors]

palette6 = glasbey.extend_palette(hex_samples, 6, colorblind_safe=True)
palette6[4], palette6[5] = palette6[5], palette6[4]
palette6[0], palette6[3] = palette6[3], palette6[0]

sns.palplot(palette6)
print(palette6)
```
